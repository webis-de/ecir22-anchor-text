# Analysis of different types of text

```python
df_term_count = pd.read_json('/mnt/ceph/storage/data-in-progress/data-teaching/theses/wstud-thesis-probst/evaluation-data/ms-marco-term-count-per-field.json', lines=True)
df_term_count = df_term_count.groupby(['kind', 'distinct_terms']).size().reset_index(name='counts')
df_term_count
```

```python
kind_to_total_term_counts = {}
for kind in df_term_count.kind.unique():
    kind_to_total_term_counts[kind] = df_term_count[df_term_count['kind'] == kind].counts.sum()

kind_to_total_term_counts
```

```python
df_term_count['percentage'] = df_term_count.apply(lambda i: i['counts']/kind_to_total_term_counts[i['kind']], axis=1)
df_term_count
```

```python
import seaborn as sns

sns.catplot(data=df_term_count[(df_term_count['distinct_terms'] <= 10) & (df_term_count['distinct_terms'] > 0)], x='distinct_terms', y='percentage', hue='kind', kind='bar')
```
