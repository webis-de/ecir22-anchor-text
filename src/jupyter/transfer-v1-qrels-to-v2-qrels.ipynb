```python
import ir_datasets
from tqdm import tqdm
marco_v2 = ir_datasets.load('msmarco-document-v2').docs_iter()
marco_v1 = ir_datasets.load('msmarco-document').docs_store()

MANUAL_MATCHED_URLS = {
    'http://www.ikea.com/': 'http://www.ikea.com/us/en/',
    'https://www.godaddy.com/': 'https://uk.godaddy.com/',
    'https://calendar.live.com/': 'https://outlook.live.com/OWA/'
}

def document_in_marco_v2(marco_v1_id):
    target_url = marco_v1.get(marco_v1_id).url
    if target_url in MANUAL_MATCHED_URLS:
        target_url = MANUAL_MATCHED_URLS[target_url]
    
    if target_url not in url_to_doc_id:
        raise ValueError('Could not handle for "' + target_url + '".')
    
    return url_to_doc_id[target_url]

doc_id_to_url = {doc.doc_id:doc.url for doc in tqdm(marco_v2)}
url_to_doc_id = {v:k for k,v in doc_id_to_url.items()}
assert len(url_to_doc_id) == len(doc_id_to_url)
```

```python
FIELD_DOC_ID = 2
QREL_DIR = '../../Data/navigational-topics-and-qrels-ms-marco-v1/'
QRELS = {
    'qrels.msmarco-entrypage-popular2.txt': 'qrels.msmarco-v2-entrypage-popular.txt',
    'qrels.msmarco-entrypage-random2.txt': 'qrels.msmarco-v2-entrypage-random.txt',
}

for src, target in QRELS.items():
    with open(QREL_DIR + src, 'r') as src_file, open(QREL_DIR + target, 'w') as target_file:
        for line in tqdm(src_file):
            line = line.split('\t')
            if 'popular2' in src and line[0] in ['32', '80']:
                continue
            if 'random2' in src and line[0] in ['86']:
                continue
                
            line[FIELD_DOC_ID] = document_in_marco_v2(line[FIELD_DOC_ID])
            line = '\t'.join(line)
            target_file.write(line + '\n')
```
