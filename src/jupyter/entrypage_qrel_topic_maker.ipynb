```python
import pandas as pd
import csv
```

```python
#liste mit allen ids und urls (nur abgedeckte) erstellen
#random stück für stück elemente wählen und mit query in liste eintragen

in_path = "~/ids_urls_queries-entrypage_random2.tsv"
out_path_qrels = "~/qrels.msmarco-entrypage-random2.txt"
out_path_topics = "~/topics.msmarco-entrypage-random2.txt"
```

```python
df_in = pd.read_csv(in_path, delimiter = "\t", header=0, usecols=[0, 1, 2], names=["id", "url", "query"], dtype='string')
df_in
```

```python
df_in = df_in.reset_index()
df_in = df_in.rename(columns={"index":"topic_id"})
df_in['topic_id'] = df_in.index
df_in['iter'] = '0'
df_in['relevancy'] = '1'
df_in
```

```python
df_qrels = pd.concat([df_in['topic_id'], df_in['iter'], df_in['id'], df_in['relevancy']], axis=1)
df_topics = pd.concat([df_in['topic_id'], df_in['query']], axis=1)
df_topics
```

```python
df_qrels.to_csv(out_path_qrels, index=False, header=False, sep='\t')
df_topics.to_csv(out_path_topics, index=False, header=False, sep='\t')
```
